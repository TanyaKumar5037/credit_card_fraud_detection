```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set styles
sns.set(style="whitegrid")
```

```python
# Load the dataset
df = pd.read_csv(r"\credit_card_fraud_detection\archive (10)\creditcard.csv")#copy file path
```

```python
df.shape, df.columns
```

```python
# Preview the data
df.head()
```

```python
# Summary info
df.info()
```

```python
# Distribution of fraud (1) vs legit (0)
print(df['Class'].value_counts())
sns.countplot(x='Class', data=df)
plt.title('Transaction Class Distribution')
plt.xticks([0, 1], ['Legit', 'Fraud'])
plt.show()
```

```python
df.isnull().sum()
```

```python
# Only show correlation with Class
corr = df.corr()['Class'].sort_values(ascending=False)
print(corr.head(10))  # Top correlated features
```

```python
# Full heatmap (optional â€“ takes time)
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Distribution of 'Amount' feature
plt.figure(figsize=(10,4))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()
# Distribution of 'Time' feature
plt.figure(figsize=(10,4))
sns.histplot(df['Time'], bins=50, kde=True)
plt.title('Transaction Time Distribution')
plt.show()
```

```python
# Compare mean amount for legit vs fraud
df.groupby('Class')['Amount'].mean()
```
